Simplify level of education


```python
included_values = ['EDU_AGGREGATE_LTB', 'EDU_AGGREGATE_BAS', 'EDU_AGGREGATE_INT', 'EDU_AGGREGATE_ADV']
df2_filtered = df2[df2['level_education'].isin(included_values)]
df2_filtered
```

Simplify the writing --> Dictionnary

```python
equivalence_dict = {
    'EDU_AGGREGATE_LTB': 'LESS_THAN_BASIC',
    'EDU_AGGREGATE_BAS': 'BASIC',
    'EDU_AGGREGATE_INT': 'INTERMEDATE',
    'EDU_AGGREGATE_ADV': 'ADVANCED'
}

df2_filtered['level_education_standardised'] = df2_filtered['level_education'].replace(equivalence_dict)
```

```python
df2_filtered1=df2_filtered.drop(['level_education',],axis=1)
df2_filtered1
```

Same for the Age 

```python
included_values2 = ['AGE_10YRBANDS_TOTAL','AGE_10YRBANDS_Y15-24','AGE_10YRBANDS_Y25-34','AGE_10YRBANDS_Y35-44','AGE_10YRBANDS_Y45-54','AGE_10YRBANDS_Y55-64','AGE_10YRBANDS_YGE65']
df2_filtered2 = df2_filtered1[df2_filtered1['age_group'].isin(included_values2)]
df2_filtered2
```

```python
equivalence_dict_age = {
    'AGE_10YRBANDS_Y15-24': '15_24',
    'AGE_10YRBANDS_Y25-34': '25_34',
    'AGE_10YRBANDS_Y35-44': '35_44',
    'AGE_10YRBANDS_Y45-54': '45_54',
    'AGE_10YRBANDS_Y55-64': '55-64'
}

df2_filtered2['age_group_filtered'] = df2_filtered2['age_group'].replace(equivalence_dict_age)

```
