Creation of derived data  
--> most popular job category per country

```python

# Step 1: Filter for age_group_simp == 'Total' and job_category != 'Total'
filtered_df = df[(df['age_group_simp'] == 'Total') & (df['job_category'] != 'Total')]
```

```python

total_workers = filtered_df.groupby(['country', 'year', 'job_category'], as_index=False)['number_worker'].sum()
most_popular = total_workers.loc[total_workers.groupby(['country', 'year'])['number_worker'].idxmax()]
```

```python
most_popular
```

```python
df['most_popular_job_cat'] = 'None'
```

```python
df
```

```python
# Step 3: Merge the most popular job category back into the original DataFrame
df_with_popular_category = filtered_df.merge(
    most_popular[['country', 'year', 'job_category', 'number_worker']],
    on=['country', 'year'],
    how='left',
    suffixes=('', '_most_popular')
)

# Rename the new column to 'most_popular_job_cat'
df_with_popular_category.rename(columns={'job_category_most_popular': 'most_popular_job_cat'}, inplace=True)
```

```python
df_with_popular_category.head(100)
```

```python
# Faire une jointure (merge) sur les colonnes 'country' et 'year'
df_merged = pd.merge(df, df_with_popular_category, on=['country', 'year'], how='left')

print(df_merged)
```

```python
# Initialize an empty dictionary to store the values
popular_job_dict = {}

# Loop through each row in df_with_popular_category
for idx, row in df_with_popular_category.iterrows():
    country = row['country']
    year = row['year']
    most_popular_job_cat = row['most_popular_job_cat']
    
    # Store the value in the dictionary only if it's not already present for that country and year
    if country not in popular_job_dict:
        popular_job_dict[country] = {}
    
    # Check if the year is already present and if the value is identical; only store if it's new/different
    if year not in popular_job_dict[country] or popular_job_dict[country][year] != most_popular_job_cat:
        popular_job_dict[country][year] = most_popular_job_cat

# The result is a dictionary with countries and years, storing unique most_popular_job_cat values.
popular_job_dict
```

```python
# Loop through each row in df
for idx, row in df.iterrows():
    country = row['country']
    year = row['year']
    age_group = row['age_group_simp']
    
    # Check if 'age_group_simp' is 'Total' and if the country and year exist in the dictionary
    if age_group == 'Total' and country in popular_job_dict and year in popular_job_dict[country]:
        # Update 'most_popular_job_cat' to the corresponding value in the dictionary
        df.at[idx, 'most_popular_job_cat'] = popular_job_dict[country][year]

# The 'df' dataset will now have updated 'most_popular_job_cat' values where 'age_group_simp' is 'Total'.
df
```
